Esta soluci√≥n se demora casi lo mismo, pero usa 10 veces m√°s memoria

```python
def q1_context_manager(file_path):
    with open(file_path, "r") as file:
        data = pd.read_json(file, lines=True)
    matches = data["content"].str.extractall(r"@(\S+?) ")
    sol = [(row[0], row['count']) for row in matches[0].value_counts().nlargest(10).reset_index().to_dict('records')]
    return sol
```

```python
cProfile.run(f"q1_context_manager('{file_path}')")
```
